3. Usando "information_schema", extraer los nombres de las columnas de la tabla "customers" de la BBDD "northwind".

```python
# Hacemos la conexión con el servidor
cnx = mysql.connector.connect(user='root', password='AlumnaAdalab',
                            host='127.0.0.1',
                            database='northwind')

# iniciamos el cursor
mycursor = cnx.cursor()

# Consulta SQL para obtener los nombres de las columnas de la tabla "customers"
query = ("SELECT column_name FROM information_schema.columns WHERE table_schema = 'northwind' AND table_name = 'customers'")

# Ejecutar la consulta
mycursor.execute(query)

# Recorrer los resultados e imprimir los nombres de las columnas
print("Nombres de las columnas de la tabla 'customers':")
for (column_name,) in mycursor:
    print(column_name)
```

4. Seleccionar todo de la tabla "customers" donde "country" sea "USA" o "Germany". Almacenar los resultados en una variable.

```python
# Hacemos la conexión con el servidor
cnx = mysql.connector.connect(user='root', password='AlumnaAdalab',
                            host='127.0.0.1',
                            database='northwind')

# iniciamos el cursor
mycursor = cnx.cursor()

# Consulta SQL para obtener los nombres de las columnas de la tabla "customers"
query = ("SELECT * FROM customers WHERE country='USA' OR country='Germany'")

# Ejecutar la consulta
mycursor.execute(query)

myresult = mycursor.fetchall()
```

```python
# EXTRA: Como vimos que el DataFrame no interpretaba los '' como NULL, decidimos recorrer los elementos para reemplazar esos '' por np.nan y que el convertir en Dataframe los NULL se mostrasen correctamente

nueva_lista = []

for tupla in myresult:
    
    tupla = list(tupla)
    
    for indice , elemento in enumerate(tupla):
        
        if elemento == '':
            
            tupla[indice] = np.nan
    
    nueva_lista.append(tuple(tupla))
```

5. Convertir los resultados del ejercicio anterior en un Dataframe (no od olvidéis de poner el nombre a las columnas en el orden correcto).

```python
df = pd.DataFrame(nueva_lista, columns = ['customer_id', 'company_name', 'contact_name','contact_title','address','city','region','postal_code', 'country', 'phone', 'fax'])
```
